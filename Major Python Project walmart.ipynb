### Analysis

#### A) To find out which category demands expansion according to the order priority , sales and orders

```python
#a.)
product_category_counts = df['product_category'].value_counts()
# Plotting the pie chart
plt.figure(figsize=(5, 5))
plt.pie(product_category_counts, labels=product_category_counts.index, autopct='%1.1f%%', 
startangle=140)
plt.title('Product Category-wise Orders')
plt.show()
```

```python
#b.)
columns_of_interest = ['product_category', 'order_priority']
subset_df = df[columns_of_interest]
# Create a crosstab to count occurrences of each product category and order priority
cross_tab = pd.crosstab(subset_df['product_category'], subset_df['order_priority'])
# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
cross_tab.plot(kind='bar', stacked=True)
plt.title('Product Categories by Order Priority')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Order Priority')
plt.show()
```

```python
#c.)
office_supplies_df = df[df['product_category'] == 'Office Supplies']
# Assuming 'product_sub_category' is the column you want to visualize
product_sub_category_counts = office_supplies_df['product_sub_category'].value_counts()
# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(product_sub_category_counts, labels=product_sub_category_counts.index, 
autopct='%1.1f%%', startangle=140)
plt.title('Product Sub-Category-wise Orders in Office Supplies')
plt.show()
```

#### Interpretations:

```python
#d.)
columns_of_interest = ['product_category', 'sales']
subset_df = df[columns_of_interest]
# Calculate total sales for each product category
total_sales_by_category = subset_df.groupby('product_category')['sales'].sum()
# Plotting the bar chart
plt.figure(figsize=(6, 3))
total_sales_by_category.plot(kind='bar', color='orange')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#e.)
columns_of_interest = ['product_sub_category']
subset_df = df[columns_of_interest]
# Calculate total orders for each product sub-category
total_orders_by_sub_category = subset_df['product_sub_category'].value_counts()
# Plotting the bar chart
plt.figure(figsize=(12, 8))
total_orders_by_sub_category.plot(kind='bar', color='purple')
plt.title('Total Orders by Product Sub-Category')
plt.xlabel('Product Sub-Category')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### B) To find out the solution for faster delivery and convenience in a region

```python
#a.)
columns_of_interest = ['region']
subset_df = df[columns_of_interest]
# Calculate total orders for each region
total_orders_by_region = subset_df['region'].value_counts()
# Plotting the bar chart
plt.figure(figsize=(12, 8))
total_orders_by_region.plot(kind='bar', color='coral')
plt.title('Total Orders by Region')
plt.xlabel('Region')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.show()

#b.)
columns_of_interest = ['region', 'Days_for_Shipping']
subset_df = df[columns_of_interest]
# Create a new column to categorize Days_for_Shipping
subset_df['Shipping_Category'] = pd.cut(subset_df['Days_for_Shipping'], bins=[0, 3, 6, float('inf')], 
labels=['1-3 days', '4-6 days', 'More than 6 days'], right=False)
# Create a crosstab to count occurrences of each region and shipping category
cross_tab = pd.crosstab(subset_df['region'], subset_df['Shipping_Category'])
# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
cross_tab.plot(kind='bar', stacked=True)
plt.title('Distribution of Days for Shipping by Region')
plt.xlabel('Region')
plt.ylabel('Number of Shipments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Days for Shipping')
plt.show()
```

#### Interpretations:

#### I) Analysing sales Across Age Groups

```python
#Define age groups
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
df['age_group'] = pd.cut(df['customer_age'], bins=age_bins, labels=age_labels, right=False)
 
# Calculate total sales across age groups
total_sales_by_age = df.groupby('age_group')['sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='sales', data=total_sales_by_age)
plt.title('Total Sales Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.show()
```

#### Interpretations:

#### J) Sales trend over Years & Quarter

```python
# Sales trend over Years, Quarter.
 
df['order_date'] = pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].dt.year
df['quarter'] = df['order_date'].dt.quarter
 
fig, axes = plt.subplots(2, 2, figsize=(18, 12), sharey='row')
 
for i, (ax, time_unit) in enumerate(zip(axes, ['year', 'quarter'])):
    # Line plot
    sns.lineplot(x=time_unit, y='sales', data=df.groupby(time_unit)['sales'].sum().reset_index(), marker='o', ax=ax[0])
    ax[0].set_title(f'Sales Trend Over {time_unit.capitalize()}s (Line Plot)')
    ax[0].set_xlabel(time_unit.capitalize())
    ax[0].set_ylabel('Total Sales')
 
    # Bar plot
    sns.barplot(x=time_unit, y='sales', data=df.groupby(time_unit)['sales'].sum().reset_index(), ax=ax[1])
    ax[1].set_title(f'Sales Trend Over {time_unit.capitalize()}s (Bar Plot)')
    ax[1].set_xlabel(time_unit.capitalize())
    ax[1].set_ylabel('Total Sales')
 
plt.tight_layout()
plt.show()
```

#### Interpretations:

#### K) Top Selling Cities, States

```python
if 'city' in df.columns and 'state' in df.columns:

    # Top Selling Cities
    top_selling_cities = df.groupby('city')['sales'].sum().sort_values(ascending=False).head(10)

    # Top Selling States
    top_selling_states = df.groupby('state')['sales'].sum().sort_values(ascending=False).head(10)

    # Plotting
    fig, axes = plt.subplots(2, 1, figsize=(12, 12))

    # Top Selling Cities
    axes[0].bar(top_selling_cities.index, top_selling_cities.values)
    axes[0].set_title('Top Selling Cities')
    axes[0].set_ylabel('Total Sales')

    # Top Selling States
    axes[1].bar(top_selling_states.index, top_selling_states.values)
    axes[1].set_title('Top Selling States')
    axes[1].set_ylabel('Total Sales')

    plt.tight_layout()
    plt.show()

else:
    print("Error: 'city' and 'state' columns not found in the DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame
# Replace 'your_file.csv' with the actual file name or provide the DataFrame in another way
# df = pd.read_csv('your_file.csv')

# Assuming 'order_date' is in datetime format
df['year'] = df['order_date'].dt.year

# Yearly sales data
yearly_sales = df.groupby(['year'])['sales'].agg(['sum']).reset_index()

# Filter data for the years 2019, 2020, 2021, 2022
filtered_data = yearly_sales[yearly_sales['year'].isin([2019, 2020, 2021, 2022])]

# Prepare data for linear regression
X = filtered_data[['year']]
y = filtered_data['sum']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict sales for the year 2023
predicted_2023 = model.predict([[2023]])

# Plotting the trend line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Actual Sales')
plt.plot([2019, 2020, 2021, 2022, 2023], [model.predict([[year]])[0] for year in [2019, 2020, 2021, 2022, 2023]], label='Trend Line', color='red')
plt.title('Yearly Sales Trend with Predictions for 2023')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend()
plt.show()

# Display the predicted sales for 2023
print(f'Predicted Sales for 2023: {predicted_2023[0]:.2f}')
```

#### l) Month-wise Sales Comparison Over Different Years

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month

# Monthly sales data
monthly_sales = df.groupby(['year', 'month'])['sales'].sum().reset_index()

# Pivot the data for better plotting
pivot_sales = monthly_sales.pivot(index='month', columns='year', values='sales')

# Plotting the line graph
plt.figure(figsize=(12, 8))
plt.plot(pivot_sales, marker='o')
plt.title('Month-wise Sales Comparison Over Different Years')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(pivot_sales.columns, title='Year', loc='upper right')
plt.grid(True)
plt.show()
```
